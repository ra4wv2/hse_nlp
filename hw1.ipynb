```python
import RAKE
from keybert import KeyBERT
from summa import keywords
import pandas as pd
import nltk
import spacy
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

# 1. Data

> 4 articles from towardsdatascience.com



```python
articles = {}
keys = {}
my_keys = {}

for i in range(1, 5):
  with open(f'data/text{str(i)}.txt', 'r', encoding='utf-8') as text_file:
    articles[i] = text_file.read()
  with open(f'data/key{str(i)}.txt', 'r', encoding='utf-8') as keys_file:
    keys[i] = keys_file.read().lower().split('\n')
  with open(f'data/mykey{str(i)}.txt', 'r', encoding='utf-8') as mykeys_file:
    my_keys[i] = mykeys_file.read().lower().split('\n')
```

# 2. Keywords



> sum of keywords for the reference



```python
article = []
sum_ = []
intersection = []
original_num = []
my_num = []
for k in keys:
  article.append(k)
  sum_.append(len(set(keys[k]+my_keys[k])))
  intersection.append(len(set(keys[k]) & set(my_keys[k])))
  original_num.append(len(keys[k]))
  my_num.append(len(my_keys[k]))

pd.DataFrame({'article':article, 'original_highlightes': original_num, 'my_highlights': my_num, 'sum': sum_, 'intersection': intersection})
```

```python
for k in keys:
  keys[k]=list(set(keys[k]+my_keys[k]))
```

# 3. RAKE, TextRank, KeyBERT

```python
results_rake = {}
results_textrank = {}
results_bert = {}
```

```python
swords = stopwords.words('english')
```

**RAKE**

```python
rake = RAKE.Rake(swords)
```

```python
for k in articles:
    results_rake[k] = rake.run(articles[k], maxWords=1, minFrequency=2)
```

**TextRank**

```python
for k in articles:
    results_textrank[k] = keywords.keywords(articles[k], language='english', additional_stopwords=swords, scores=True)
```

**KeyBERT**

```python
kw_model = KeyBERT('clips/mfaq')
```

```python
for k in articles:
    results_bert[k] = kw_model.extract_keywords(articles[k], keyphrase_ngram_range=(1, 2), stop_words=swords)
```

# 4. Keywords models

```python
sp = spacy.load('en_core_web_sm')
```

```python
keys_models = {}

for k in keys:
  all_mdls = []

  for key in keys[k]:
    word = sp(key)
    model_kw = [word[i].pos_ for i in range(len(key.split(' ')))]
    if model_kw not in all_mdls:
      all_mdls.append(model_kw)

  keys_models[k] = all_mdls
```

```python
def check_keywords(kwords, models):
  filltered = {n:[]for n in kwords}

  for k in kwords:
    for key in kwords[k]:
      word = sp(key[0])
      model_kw = []
      for i in range(len(key[0].split(' '))):
        model_kw.append(word[i].pos_)
      if model_kw in models[k]:
        filltered[k].append(key)
        
  return filltered
```

```python
filtered_bert = check_keywords(results_bert, keys_models)
filtered_rake = check_keywords(results_rake, keys_models)
filtered_textrank = check_keywords(results_textrank, keys_models)
```

# 5. Precision, recall, F-score

```python
def count_final_score(real_kws, pred_kws):
  scores = {}

  for k in range(1, 5):
    tp = sum([1 for word in pred_kws[k] if word[0] in real_kws[k]])
    fn = sum([1 for word in real_kws[k] if word[0] not in pred_kws[k]])
    fp = int(sum([1 for word in pred_kws[k] if word[0] not in real_kws[k]]))

    if tp != 0:
      precision = tp / (tp + fp)
      recall = tp / (tp + fn)
      f_score = 2 * precision * recall / (precision + recall)
    else:
      precision = 0
      recall = 0
      f_score = 0

    scores[k] = (round(precision, 5), round(recall, 5), round(f_score, 5))

  return scores
```

### Scores for each article

```python
mthds = []
filters = []
articles = []
p = []
r = []
f = []

methods = ['rake', 'textrank', 'bert']

for m in methods:
  for i in range(2):
    if i == 0:
      dt = count_final_score(keys, globals()['results_'+m])
    else:
      dt = count_final_score(keys, globals()['filtered_'+m])
    for k in dt:
      mthds.append(m)
      filters.append(i)
      articles.append(str(k))
      p.append(dt[k][0])
      r.append(dt[k][1])
      f.append(dt[k][2])
```

```python
df = pd.DataFrame({'method':mthds, 'filter':filters, 'article':articles, 'Presicion':p, 'Recall':r, 'F-score':f})
```
